PPTX -> TEXT

```python
from pptx import Presentation
def extract_pptx(file_path):
    prs = Presentation(file_path)
    documents=[]
    ppt_name = os.path.basename(file_path)
    for slide_no, slide in enumerate(prs.slides):
        slide_text = ""
        for shape in slide.shapes:
            if hasattr(shape , "text"):
                slide_text += shape.text + "\n"
        if slide_text.strip():
            documents.append(
                {
                    "content":
                        f"Presentation: {ppt_name}\n"
                        f"Slide Number: {slide_no+1}\n"
                        f"{slide_text}",
                    "type": "text",
                    "slide": slide_no + 1
                }
            )
    return documents
```
